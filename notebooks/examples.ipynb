```python
nato_words = [
    'alpha', 'bravo', 'charlie', 'delta', 'echo',
    'foxtrot', 'golf', 'hotel', 'india', 'juliet',
    'kilo', 'lima', 'mike', 'november', 'oscar',
    'papa', 'quebec', 'romeo', 'sierra', 'tango',
    'uniform', 'victor', 'whiskey', 'x-ray', 'yankee', 'zulu'
]

import numpy as np
from sklearn.manifold import MDS
import matplotlib.pyplot as plt

# Create distance matrix
n_words = len(nato_words)
distances = np.zeros((n_words, n_words))

# Compute pairwise distances
aligner = WordAligner(weighted=True, gap_penalty=1.5)
for i in range(n_words):
    for j in range(i+1, n_words):
        _, _, dist = aligner.compare_words(nato_words[i], nato_words[j])
        distances[i,j] = dist
        distances[j,i] = dist # Matrix is symmetric

# Perform MDS
mds = MDS(n_components=2, dissimilarity='precomputed', random_state=42)
mds.fit(distances)
coords = mds.transform(distances)

# Plot results
plt.figure(figsize=(12, 8))
plt.scatter(coords[:, 0], coords[:, 1])

# Add labels for each point
for i, word in enumerate(nato_words):
    plt.annotate(word, (coords[i, 0], coords[i, 1]))

# Perform MDS
mds = MDS(n_components=2, dissimilarity='precomputed', random_state=42)
mds.fit(distances)
coords = mds.fit_transform(distances)

# Plot results
plt.figure(figsize=(6, 6))
plt.scatter(coords[:, 0], coords[:, 1])

plt.title('MDS visualization of NATO phonetic alphabet word distances')
plt.xlabel('First dimension')
plt.ylabel('Second dimension')
plt.show()
```
